count plot of sex

```python
data = {
    'sex':[1, 1, 0, 1, 0, 1, 1, 1, 1, 0]
}
df=pd.DataFrame(data)
sns.countplot(data=df,x='sex')
plt.title('count plot of sex')
plt.xlabel('sex')
plt.ylabel('count')
plt.show()
```

pie chart

```python
data={
    'sex'=[1,1,0,1,0,1,1,1,1,0]
}

df = pd.DataFrame(data)

df['sex'].value_counts().plot(kind='pie', autopct='%0.2f', labels=['Male', 'Female'],colors=['lightblue', 'pink'])
plt.title('Proportion of Sex')
plt.show()
```

```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
```

box plot

```python
age = [21, 23, 30, 34, 35, 36, 39, 50, 64, 68]
diabp = [100, 85, 73, 80, 90, 109, 90, 117, 96, 96]

plt.bar(range(len(age)), diabp, color='skyblue')

plt.title("Diastolic Blood Pressure by Age")
plt.ylabel("Diastolic Blood Pressure (DiaBP)")
plt.xlabel("Individuals (Sorted by Age)")

plt.xticks(range(len(age)), age)

plt.show()
```

line chart

```python
data = {
    "age": [64, 21, 30, 35, 39, 68, 50, 23, 34, 36],
    "SysBP": [141, 109, 112, 106, 140, 115, 125, 111, 144, 113]
}

df = pd.DataFrame(data)

age = df["age"].tolist()
sysbp = df["SysBP"].tolist()

plt.plot(sysbp, age, color='blue', marker='o', linestyle='dashed')

plt.title("Age vs Systolic Blood Pressure")
plt.xlabel("Systolic Blood Pressure (SysBP)")
plt.ylabel("Age (Years)")

plt.show()
```
